```python
import pandas as pd
data = pd.read_csv('smata.csv')
data
```

```python
from math import floor

data = data.groupby('question_id').agg({
    'id':'last',
    'times_option_changed': 'sum',
    'is_correct': 'last',
    'user_confidence': lambda x: floor(x.mean()),
    'time_spent': 'sum'
}).reset_index()

data = data[['id', 'times_option_changed', 'is_correct', 'question_id', 'user_confidence', 'time_spent']]
data
```

```python
import django
from django.conf import settings

import os
import sqlite3

# Set the DJANGO_SETTINGS_MODULE environment variable
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'ils.settings')

# Initialize Django
django.setup()

db_path = os.path.join(settings.BASE_DIR, 'db.sqlite3')
conn = sqlite3.connect(db_path)

table_name = 'quiz_question'
query = f"SELECT * FROM {table_name}"
df_question = pd.read_sql_query(query, conn)
df_question

question_cognitive_ability = dict(zip(df_question['id'], df_question['cognitive_ability']))

data['cognitive_ability'] = data['question_id'].map(question_cognitive_ability)
taxonomy_level_mapping = {'Level_1': 1, 'Level_2': 2, 'Level_3': 3}
data['cognitive_ability'] = data['cognitive_ability'].map(taxonomy_level_mapping)
data
```

```python
X = data[['time_spent', 'times_option_changed', 'is_correct', 'cognitive_ability']]

import joblib
model = joblib.load('confidence_predictor_model.pkl')

predicted_confidence = model.predict(X)
data['predicted_confidence'] = predicted_confidence
data
```
